# Data Loading

[Ref] https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

pd.set_option("display.max_columns", 100)
sns.set_style("whitegrid")
```

```python
_data = load_iris()
data = _data['data']
feature_names = _data['feature_names']

df = pd.DataFrame(data, columns=feature_names)
df['target'] = _data['target']

df.head()
```

# Preprocessing

```python
# target 재정의
df['target'] = df['target'].map(
    {
        0: 'Iris-Setosa',
        1: 'Iris-Versicolour',
        2: 'Iris-Virginica',
    }
)

df.head()
```

```python
df['target'].unique()
```

# EDA

```python
df.groupby('target').describe()
```

```python
sns.boxplot(
    x='target',
    y='petal length (cm)',
    data=df,
)
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(221)
sns.boxplot(x="target", y="petal length (cm)", data=df)

plt.subplot(222)
sns.boxplot(x="target", y="petal width (cm)", data=df)

plt.show()
```

```python
_df_skew = df.groupby("target").skew()

_df_skew.T.plot(kind="bar", grid=True)
plt.title("Skewness by Feature with IRIS")
plt.ylabel("Skewness")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## StandardScaler 

- (가정) 데이터셋의 컬럼별로 각각 단위가 다르고, Target 범주별 값의 분포 차이가 상이함

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
```

```python
_df_scaled = scaler.fit_transform(df)
```

```python
_df = df.iloc[:, :-1]
_df.head()
```

```python
_df_scaled = scaler.fit_transform(_df)
df_scaled = pd.DataFrame(_df_scaled, columns=_df.columns)
df_scaled["target"] = df["target"]

df_scaled.head()
```

```python
plt.figure(figsize=(10, 10))
plt.title("Standard Scaler")

plt.subplot(221)
sns.boxplot(x="target", y="petal length (cm)", data=df)

plt.subplot(222)
sns.boxplot(x="target", y="petal width (cm)", data=df)

plt.subplot(223)
sns.boxplot(x="target", y="petal length (cm)", data=df_scaled)

plt.subplot(224)
sns.boxplot(x="target", y="petal width (cm)", data=df_scaled)

plt.show()
```

## Min-Max Scaler

- (가정) 데이터셋의 컬럼별로 각각 단위가 다르고, Target 범주별 값의 분포 차이가 상이함

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
_df_scaled = scaler.fit_transform(_df)
df_scaled = pd.DataFrame(_df_scaled, columns=_df.columns)
df_scaled["target"] = df["target"]

df_scaled.head()
```

```python
plt.figure(figsize=(10, 10))
plt.title("Min-Max Scaler")

plt.subplot(221)
sns.boxplot(x="target", y="petal length (cm)", data=df)

plt.subplot(222)
sns.boxplot(x="target", y="petal width (cm)", data=df)

plt.subplot(223)
sns.boxplot(x="target", y="petal length (cm)", data=df_scaled)

plt.subplot(224)
sns.boxplot(x="target", y="petal width (cm)", data=df_scaled)

plt.show()
```
