```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

pd.set_option("display.max_columns", 100)
sns.set_style("whitegrid")
```

# Data Loading

[Ref] https://en.wikipedia.org/wiki/Iris_flower_data_set

```python
_data = load_iris() #_로 데이터 프레임의 완성정도를 표현 _a -> a -> a_ 수
_data.keys()
```

```python
data = _data['data']
feature_names = _data['feature_names']

df = pd.DataFrame(data, columns=feature_names)
df['target'] = _data['target']

df.head()
```

# Preprocessing

```python
df.shape
```

```python
# df.isnull()
df.isnull().sum()
```

```python
df.describe()
```

```python
# target 재정의
df['target'] = df['target'].map(
    {
        0: 'Iris-Setosa',
        1: 'Iris-Versicolour',
        2: 'Iris-Virginica',
    }
)

df.head()
```

```python
df['target'].unique()
```

```python
df.describe()
```

# EDA

## Descriptive Statistics

```python
df.groupby('target').describe()
```

```python
sns.pairplot(  #pairplot 큰 데이터를 처리하기에는 무거움 
    df,
    hue='target',
    size=3,  #점들의 사이즈 
    diag_kind='kde',
)
```

```python
df.groupby('target').skew() #skew는 기울어진 정도, 함수를 필요로 함
```

### skew

```python
_df_skew = df.groupby('target').skew()

_df_skew.T.plot(kind='bar', grid=True)
plt.title('Skewness by Feature with IRIS')
plt.ylabel('Skewness')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### kurtosis

```python
df.groupby('target').kurtosis()  # 뾰족한 정도 
```

```python
df.groupby("target").apply(
    lambda group: group.iloc[:, :-1].kurtosis()
)
```

```python
_df_kurt = df.groupby("target").apply(lambda group: group.iloc[:, :-1].kurtosis())

_df_kurt.T.plot(kind='bar', grid=True)
plt.title('Kurtosis by Feature with IRIS')
plt.ylabel('kurtosis')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
